```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
def faturamento(loja):
  faturamento = loja["Preço"].sum()
  return faturamento

```

```python
print("Loja 1:")
print("faturamento -> ",faturamento(loja))
print("------------------------------")
print("Loja 2:")
print("faturamento -> ", faturamento(loja2))
print("------------------------------")
print("Loja 3:")
print("faturamento -> ",faturamento(loja3))
print("------------------------------")
print("Loja 4:")
print("faturamento -> ",faturamento(loja4))
```

# 2. Vendas por Categoria


```python
def vendaCategoria(loja):

    qtd = loja["Categoria do Produto"].value_counts()

    return qtd

```

```python
print("Loja 1:")
print(vendaCategoria(loja))
print("------------------------------")
print("Loja 2:")
print(vendaCategoria(loja2))
print("------------------------------")
print("Loja 3:")
print(vendaCategoria(loja3))
print("------------------------------")
print("Loja 4:")
print(vendaCategoria(loja4))
```

# 3. Média de Avaliação das Lojas

```python
def avaliacao(loja):
  qtd = len(loja['Avaliação da compra'])
  soma = loja['Avaliação da compra'].sum()
  media = soma/qtd
  mediaR = round(media, 2)

  return mediaR
```

```python
print("Loja 1:")
print("média das avaliações dos produtos -> ",avaliacao(loja))
print("------------------------------")
print("Loja 2:")
print("média das avaliações dos produtos -> ", avaliacao(loja2))
print("------------------------------")
print("Loja 3:")
print("média das avaliações dos produtos -> ",avaliacao(loja3))
print("------------------------------")
print("Loja 4:")
print("média das avaliações dos produtos -> ",avaliacao(loja4))
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt

def produtos_mais_vendidos(loja):
    contagem = loja["Produto"].value_counts()
    mais_vendidos = contagem.head(5)

    return mais_vendidos

def produtos_menos_vendidos(loja):
    contagem = loja["Produto"].value_counts()
    menos_vendidos = contagem.tail(5)

    return menos_vendidos


def visualizar_vendas(loja, nome_loja="Loja"):
    contagem = loja["Produto"].value_counts()
    mais_vendidos = contagem.head(5)
    menos_vendidos = contagem.tail(5)

    # Gráfico: Top 5 mais vendidos
    plt.figure(figsize=(5,4))
    mais_vendidos.plot(kind="bar", color="blue", title=f"{nome_loja} - Top 5 Mais Vendidos")
    plt.ylabel("Quantidade Vendida")
    plt.xlabel("Produto")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
print("Loja 1:")
print("--Produtos mais vendidos--")
print(produtos_mais_vendidos(loja))
print(" ")
print("--Produtos menos vendidos--")
print(produtos_menos_vendidos(loja))
print(" ")
print("--Grafico dos produtos mais vendidos--")
print(" ")
visualizar_vendas(loja, "Loja 1")
print("------------------------------")
print("Loja 2:")
print("--Produtos mais vendidos--")
print(produtos_mais_vendidos(loja2))
print(" ")
print("--Produtos menos vendidos--")
print(produtos_menos_vendidos(loja2))
print(" ")
print("--Grafico dos produtos mais vendidos--")
print(" ")
visualizar_vendas(loja2, "Loja 2")
print("------------------------------")
print("Loja 3:")
print("--Produtos mais vendidos--")
print(produtos_mais_vendidos(loja3))
print(" ")
print("--Produtos menos vendidos--")
print(produtos_menos_vendidos(loja3))
print(" ")
print("--Grafico dos produtos mais vendidos--")
print(" ")
visualizar_vendas(loja3, "Loja 3")
print("------------------------------")
print("Loja 4:")
print("--Produtos mais vendidos--")
print(produtos_mais_vendidos(loja4))
print(" ")
print("--Produtos menos vendidos--")
print(produtos_menos_vendidos(loja4))
print(" ")
print("--Grafico dos produtos mais vendidos--")
print(" ")
visualizar_vendas(loja4, "Loja 4")
```

# 5. Frete Médio por Loja


```python
def frete(loja):
  qtd = len(loja['Frete'])
  soma = loja['Frete'].sum()
  media = soma/qtd
  mediaR = round(media, 2)

  return mediaR
```

```python
print("Loja 1:")
print("média do frete -> R$", frete(loja))
print("------------------------------")

print("Loja 2:")
print("média do frete -> R$", frete(loja2))
print("------------------------------")

print("Loja 3:")
print("média do frete -> R$", frete(loja3))
print("------------------------------")

print("Loja 4:")
print("média do frete -> R$", frete(loja4))
```

# 6. Gráficos

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


plt.style.use('seaborn-v0_8-muted')

cores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [1534509.12, 1488459.06, 1464025.03, 1384497.58]
fretes = [34.69, 33.62, 33.07, 31.28]
avaliacoes = [3.98, 4.04, 4.05, 4.0]

# Gráfico 1: Faturamento e Frete
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

ax1.bar(lojas, faturamentos, color=cores)
ax1.set_title('Faturamento por Loja', fontsize=10)
ax1.set_ylabel('R$', fontsize=9)
ax1.tick_params(axis='x', labelrotation=0)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'R${x/1000:.0f}K'))

for i, v in enumerate(faturamentos):
    ax1.text(i, v + 20000, f'R${v/1000:.0f}K', ha='center', fontsize=8)

ax2.bar(lojas, fretes, color=cores)
ax2.set_title('Frete Médio por Loja', fontsize=10)
ax2.set_ylabel('R$', fontsize=9)

for i, v in enumerate(fretes):
    ax2.text(i, v + 0.2, f'R${v:.2f}', ha='center', fontsize=8)

plt.tight_layout()
plt.show()


# Gráfico 2: Categorias - Pizza
categorias = {
    'Móveis': 465,
    'Eletrônicos': 448,
    'Brinquedos': 324,
    'Eletrodomésticos': 312,
    'Esporte e Lazer': 284
}

plt.figure(figsize=(5, 4))
plt.pie(categorias.values(), labels=categorias.keys(), autopct='%1.1f%%',
        startangle=90, colors=sns.color_palette("Set3"), textprops={'fontsize': 9})
plt.title('Distribuição por Categoria', fontsize=10, pad=10)
plt.axis('equal')
plt.tight_layout()
plt.show()


# Gráfico 3: Faturamento vs Avaliação
fig, ax1 = plt.subplots(figsize=(7, 3.5))

ax1.plot(lojas, faturamentos, color='#2ca02c', marker='o', linewidth=2, label='Faturamento')
ax1.set_ylabel('Faturamento (R$)', color='#2ca02c', fontsize=9)
ax1.tick_params(axis='y', colors='#2ca02c')
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'R${x/1000:.0f}K'))
ax1.set_ylim(1350000, 1600000)

ax2 = ax1.twinx()
ax2.plot(lojas, avaliacoes, color='#d62728', marker='s', linewidth=2, label='Avaliação Média')
ax2.set_ylabel('Avaliação (1–5)', color='#d62728', fontsize=9)
ax2.tick_params(axis='y', colors='#d62728')
ax2.set_ylim(3.7, 4.2)

ax1.grid(linestyle='--', alpha=0.5)
plt.title('Faturamento vs Satisfação', fontsize=10, pad=10)


handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles1 + handles2, labels1 + labels2, loc='upper left', fontsize=8)

plt.tight_layout()
plt.show()
```

Gráfico 1:
Este gráfico de barras paralelas revela uma relação inversa entre faturamento e custo de frete. A Loja 1, que possui o maior faturamento (1,53M), também apresenta o frete mais caro (34,69), enquanto a Loja 4, com o menor faturamento (1,38M), tem o frete mais barato (31,28). Isso sugere que:

O frete alto não impede vendas (Loja 1 lidera em receita), mas pode indicar um público menos sensível a custos.

A Loja 4, mesmo com frete competitivo, não converte em maior faturamento, sugerindo outros fatores em jogo (como mix de produtos ou localização).

Gráfico 2:
Móveis (23,4%) e Eletrônicos (22,5%) dominam as vendas, somando quase 46% do total. As categorias menos representadas (como Instrumentos Musicais e Livros) têm margens de contribuição potencialmente maiores, mas baixo volume.

Padrão consistente: Todas as lojas seguem essa distribuição, indicando que a demanda por móveis/eletrônicos é estável.

Gráfico 3:
A Loja 3 se destaca com a melhor avaliação (4.05) e faturamento médio (R$ 1,46M), enquanto a Loja 1, líder em faturamento, tem a pior nota (3.98). Isso indica que:

Satisfação não depende só de preço: A Loja 4 tem frete barato e avaliação média (4.0), mas não converte em receita.

A Loja 1 foca em volume às custas da experiência, enquanto a Loja 3 mantém um equilíbrio ideal entre os dois fatores.

# Relatório final

Após analisar o desempenho das quatro lojas em termos de faturamento, mix de produtos, avaliações de clientes, perfil de vendas por item e custo médio de frete, concluímos que a melhor opção para o Senhor João é vender a **Loja 4**. Apesar de apresentar o frete mais barato, em média R\$ 31,28, essa redução de custo não se traduziu em vantagem competitiva: a Loja 4 apresentou o menor faturamento, com R\$ 1.384.497,58, e mantinha uma avaliação de clientes apenas mediana, nota 4,00. Em contraste, a liderança em receita ficou com a Loja 1 (R\$ 1.534.509,12), ainda que ela praticasse o frete mais alto (R\$ 34,69), indicando que seu público é menos sensível a custos de envio e confia no valor oferecido. A Loja 3, por sua vez, aliou bom desempenho financeiro (R\$ 1.464.025,03) à melhor avaliação média (4,05), demonstrando equilíbrio exemplar entre volume de vendas e satisfação do cliente. As categorias de produtos mais vendidas – móveis e eletrônicos, com cerca de 23,4 % e 22,5 % das transações, respectivamente – apresentaram padrões de demanda semelhantes em todas as lojas, enquanto nichos de maior margem, como instrumentos musicais e livros, ficaram em torno de 9–10 % das vendas, revelando oportunidades de diversificação. Além disso, os relatórios de top 5 e bottom 5 produtos destacaram itens de baixa rotatividade que ocupam espaço de estoque sem contribuir significativamente para o faturamento, sugerindo que recursos destinados à Loja 4 seriam mais bem aproveitados fortalecendo o mix de itens mais populares e testando campanhas para nichos de alta margem nas demais unidades.
Com a venda da Loja 4, o Senhor João poderá realocar o capital obtido em melhorias de experiência e marketing para as lojas mais sólidas, especialmente a Loja 3, além de investir em iniciativas de alto retorno, otimizando seu portfólio e impulsionando ainda mais o desempenho geral do negócio.
